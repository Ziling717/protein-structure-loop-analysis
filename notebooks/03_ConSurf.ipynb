#### 3.2.1 Conservation scores of loop region
The layers for assigning grades are as follows.  (9 - conserved, 1 - variable)
- from -1.208 to -0.940 the grade is 9
- from -0.940 to -0.671 the grade is 8
- from -0.671 to -0.403 the grade is 7
- from -0.403 to -0.134 the grade is 6
- from -0.134 to  0.300 the grade is 5
- from  0.300 to  0.899 the grade is 4
- from  0.899 to  1.498 the grade is 3
- from  1.498 to  2.097 the grade is 2
- from  2.097 to  2.696 the grade is 1

```python
# BfIMTD
import matplotlib.pyplot as plt

residues = list(range(392, 403))
grades = [0.219, -0.213, 0.999, 2.393, 2.076, -0.994, 0.410, 2.339, -0.314, 0.839, 0.694] 

plt.figure(figsize=(10, 4))
plt.bar(residues, grades, color='cornflowerblue')
plt.xticks(residues)

plt.xlabel("Residue Number")
plt.ylabel("Conservation Grade (1â€“9)")
plt.title("ConSurf Conservation Scores of BfIMTD Loop Region (392â€“402)")
plt.ylim(-2, 3)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### 3.3 Results and Conclusions

### Functional convergence in a divergent loop region

1. Despite overall sequence divergence in the loop region, all three GH49 dextranases analyzed (BfIMTD, CcDex, MrDex) share a highly conserved **glutamic acid (E)** residue at the center of the loop:

- **BfIMTD**: E397 â€” Grade 9 (ConSurf), exposed (`e`), predicted functional (`f`)
- **CcDex**: E397 â€” Grade 9, `e`, `f`
- **MrDex**: E386 â€” Grade 9, `e`, `f`

2. All three loops show the same **exposure pattern**: `eeeeeebebe`, with the conserved Glu centrally located.

This supports that **functional dynamics can be conserved even in sequence-divergent structural regions**.

3. The loop regions of all three enzymes appear to be evolutionarily flexible and structurally dynamic as a whole.
   
   However, each loop contains a centrally located, highly conserved glutamic acid residue, which is consistently predicted to be functionally important.

   <span style="color:green">This suggests that, despite overall sequence divergence, the loop may play a conserved functional role possibly in ligand recognition, binding, or conformational gating or allosteric regulation during catalysis.</span>
